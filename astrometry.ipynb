```python
objects=glob('./CoAdds/*')
print(objects[2:])
ims=glob(objects[0]+'/*.fits')
print(ims)
```

```python
for obj in objects:
    obj_name=obj.split('/')[-1]
    #print(obj_name)
    images=glob(obj+'/*.fits')
    for image in images:
        #print(image.lstrip('./'))
        filt_name=image.split('/')[-1]
        #print(filt_name)
        new_loc=f'WCS_Solved/{obj_name}/{obj_name}_{filt_name}'
        os.system(f'python astrometry.net/astrometry/net/client/client.py -u {image.lstrip('./')} -k {key} --newfits {new_loc}')
```
