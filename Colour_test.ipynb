```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from glob import glob
from astropy.visualization import make_lupton_rgb
plt.rcParams["figure.figsize"] = (13,8)
```

```python
ngc3705=glob('./Processed/NGC3705/*.fits')
print(ngc3705)
```

```python
im_data=fits.open(ngc3705[5])[0].data
print(ngc3705[5])
plt.imshow(im_data,vmin=20,vmax=200,cmap='inferno')
#plt.xlim(500,800)
#plt.ylim(700,1000)
```

```python
g=np.sum([fits.open(el)[0].data for el in ngc3705[0:3]],axis=0)
np.shape(g)
```

```python
g=np.sum([fits.open(el)[0].data for el in ngc3705[0:3]],axis=0)
b=np.sum([fits.open(el)[0].data for el in ngc3705[3:6]],axis=0)
r=np.sum([fits.open(el)[0].data for el in ngc3705[6:]],axis=0)
```

```python
rgb = make_lupton_rgb(r, 2.5*g, 10*b, Q=10, minimum=2000, stretch=0.2, filename="./ngc3705.jpeg")
plt.imshow(rgb, origin='lower')
#plt.xlim(400,800)
#plt.ylim(650,1100)
```
